```python
url='https://www.zeebiz.com/trending/news-animal-box-office-collection-film-nears-rs-900-core-mark-269112'
```

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
html_content=requests.get(url).text
```

```python
html_content
```

```python
soup=BeautifulSoup(html_content,'lxml')
```

```python
body=soup.body.text
```

```python
body=body.replace('\n'," ")
body=body.replace('\t'," ")
body=body.replace('\xa0'," ")
```

```python
body.strip()
```

```python
sos='Animal Box Office collection, film nears Rs 900 core mark'
eos='Triptii Dimri in significant roles'
```

```python
body.index(sos)
```

```python
body.index(eos)
```

```python
len(eos)
```

```python
body[2900]
```

```python
si=body.index(sos)
```

```python
ei=body.index(eos)+len(eos)
```

```python
body=body[si:ei]
```

```python
text=NER(body)
```

```python
displacy.render(text,style='ent',jupyter=True)
```

```python
per=set()

for i in text.ents:
  if i.label_=='PERSON':
    per.add(i)
```
